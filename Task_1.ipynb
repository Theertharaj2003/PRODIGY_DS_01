```python
import pandas as pd
df_male = pd.read_csv('/content/API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_2033140.csv', delimiter=',', on_bad_lines='skip')
df_male.head()
```

```python
import pandas as pd
df_female = pd.read_csv('/content/API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_2031266.csv', delimiter=',', on_bad_lines='skip')
df_female.head()
```

```python
data_2021 = df[['Country Name', '2021']]

data_2021 = data_2021.dropna(subset=['2021'])

data_2021 = data_2021.sort_values(by='2021',ascending=True)

data_2021 = data_2021.head(10)

top_10 = data_2021

plt.figure(figsize=(10, 6))
plt.bar(top_10['Country Name'], top_10['2021'])
plt.xlabel('Country')
plt.ylabel('Total Population in 2021')
plt.title('Population in 2021 at some countries ')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data_male_2021 = df_male[['Country Name', '2021']]
data_male_2021 = data_male_2021.dropna(subset=['2021'])
data_male_2021 = data_male_2021.sort_values(by='2021',ascending=True)
data_male_2021 = data_male_2021.head(15)


data_female_2021 = df_female[['Country Name', '2021']]
data_female_2021 = data_female_2021.dropna(subset=['2021'])
data_female_2021 = data_female_2021.sort_values(by='2021',ascending=True)
data_female_2021 = data_female_2021.head(15)

plt.figure(figsize=(12, 6))

plt.bar(data_male_2021['Country Name'], data_male_2021['2021'], label='Male')
plt.bar(data_female_2021['Country Name'], data_female_2021['2021'], bottom=data_male_2021['2021'], label='Female')

plt.xlabel('Country')
plt.ylabel('Population in 2021')
plt.title('Male and Female Population in 2021 at some countries')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
